```python
import pandas as pd
```

```python
df = pd.read_csv("./product.csv")
```

```python
def clean_text(row):
    _id = row['_id'] if pd.notna(row['_id']) else ""
    nameProduct = row['nameProduct'] if pd.notna(row['nameProduct']) else ""
    category = row['category'] if pd.notna(row['category']) else ""
    price = row['price'] if pd.notna(row['price']) else ""
    return f"{_id} {nameProduct} {category} {price}".lower()

df['clean_text'] = df.apply(clean_text, axis=1)
df = df.dropna(subset=['clean_text'])
```

```python
df['clean_text']
```

```python
df.shape
```

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('all-MiniLM-L6-v2')
df['embeddings'] = df['clean_text'].apply(lambda x: model.encode(x))
```

```python
df.to_pickle('products_embeddings.pkl')

df = pd.read_pickle('products_embeddings.pkl')
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def recommend_products(query, top_k=20):

    # Embed the query
    query_embedding = model.encode(query)

    # Compute consine similarity with all products
    df['similarity'] = df['embeddings'].apply(lambda x: cosine_similarity([query_embedding], [x]).flatten()[0])

    # sort products by similarity score
    recommendations = df.sort_values(by='similarity', ascending=False).head(top_k)
    return recommendations[['_id','nameProduct', 'category', 'price', 'images[0]', 'similarity']]
```

```python
query = "dieu hoa"
recommendations = recommend_products(query)
recommendations
```
